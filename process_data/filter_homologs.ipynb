```python
def get_repr(df, name):
    prot_id = list(df["Protein ID"])
    fasta = list(df["simple_fasta"])
    with open(name+'.fasta', "w") as fi:
        for i in range(len(prot_id)):
            fi.write(">" + prot_id[i] + "\n")
            fi.write(fasta[i] + "\n")
    subprocess.run(f"cd-hit -i {name}.fasta -o {name}cd_hit_out -c 0.5 -n 2 -d 100 -T 0", shell=True)
    
           
    
```
